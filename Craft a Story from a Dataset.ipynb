Nutrition Facts for a Starbucks Menu

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('archive\starbucks_drinkMenu_expanded.csv',index_col=False)
```

```python
print(df.head())
print(df.shape)
print(df.columns)
print(df.dtypes)
```

The concept of foods or beverages with good nutritional value is highly sought after in recent years. There are multiple trends of high-protein/low calorie/whole food ingredient hacks circulating social media. Videos explaining how to get the same satisfaction from a healthier version of one's favorite foods have become increasingly popular. There have even been Instagram reels and TikToks about how to order Starbucks drinks to get the best flavor without compromising nutrition goals. 

How much compromise do people have to actually make on their favorite Starbucks beverage? Let's investigate the average nutritional value per drink to see how it stacks up against a person's average caloric intake. Are the macronutrients as bad as we would expect? Or can a Starbucks drink actually fit reasonably withing someone's healthy eating lifestyle? 

```python
df.describe()
```

The mean calories for a drink being around 194 actually isn't that bad in respect to the fact that the mean caloric intake for an individual typically ranges from 1600-2200 for women and 2000-3200 for men. Some individuals also have higher caloric needs depending on a lot of factors including their physcial activity levels, genetics, weight, height, health conditions etc. 

The average sugar intake is pretty high at 32g and the total carbohydrates at about 129g. But let's explore the nutritional facts for the classsic drinks in particular. What if seasonal or specialty drinks carry the highest calorie content but people only purchase them a few times a year?  

```python
df['Beverage_category'].tolist()
```

```python
classic = ['Classic Espresso Drinks']
df_classic = df.loc[df['Beverage_category'].isin(classic)]
df_classic.describe()
```

Specific categories of drinks definitiely can affect the macronutrients! Mean calories decreased by about 53, carbohydrates by about 25, sugars by about 16. 

Let's visualize the percentage of a person's daily caloric intake and macronutrtients the average Starbuck's drink would comprise of. 

```python
cal = df['Calories'].mean()
carb = df[' Total Carbohydrates (g) '].mean()
sugar = df[' Sugars (g)'].mean()
protein = df[' Protein (g) '].mean()
fat = df['Trans Fat (g) '].mean() + df['Saturated Fat (g)'].mean()

df_mean_nutr = pd.DataFrame({'Calories':cal,'Carbs':carb, 'Sugars':sugar,'Protein':protein,'Fat':fat},index=[0])
df_mean_nutr
```

```python
df_chart = pd.DataFrame({'Category': ['Total Calories', 'Starbucks'],
                   'Calories': [2200, cal]})
df_chart.set_index('Category', inplace=True)
print(df_chart)

plot = df_chart.plot.pie(y='Calories', figsize=(7, 7),colors=['pink','palevioletred'])
```

```python
df_chart2 = pd.DataFrame({'Category': ['Total Carbs', 'Starbucks'],
                   'Carbs': [358, carb]})
df_chart2.set_index('Category', inplace=True)
print(df_chart2)

plot = df_chart2.plot.pie(y='Carbs', figsize=(7, 7),colors=['blanchedalmond','palevioletred'])
```

```python
df_chart3 = pd.DataFrame({'Category': ['Total Fat', 'Starbucks'],
                   'Fat': [80, fat]})
df_chart3.set_index('Category', inplace=True)
print(df_chart3)

plot = df_chart3.plot.pie(y='Fat', figsize=(7, 7),colors=['thistle','palevioletred'])
```
